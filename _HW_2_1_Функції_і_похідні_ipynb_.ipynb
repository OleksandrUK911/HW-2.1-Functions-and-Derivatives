### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Символьна змінна
x = sp.Symbol('x')

# Задана функція
f = 10 * x - 5

# Обчислення похідної
f_derivative = sp.diff(f, x)
print("Похідна функції:", f_derivative)

# Перетворення на числову функцію
f_func = sp.lambdify(x, f, 'numpy')
f_der_func = sp.lambdify(x, f_derivative, 'numpy')

# Значення x від 0 до 10
x_vals = np.linspace(0, 10, 200)
y_vals = f_func(x_vals)

# ПОПРАВКА: Похідна — сталий масив
y_der_vals = np.full_like(x_vals, f_der_func(0))

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5', linewidth=2)
plt.plot(x_vals, y_der_vals, label="f'(x) = 10", linestyle='--')
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість годин навчання (x)')
plt.ylabel('Оцінка / Швидкість зміни оцінки')
plt.legend()
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Символьна змінна
x = sp.Symbol('x')

# Функції
f1 = sp.sin(x)
f2 = sp.ln(x)

# Похідні
f1_der = sp.diff(f1, x)
f2_der = sp.diff(f2, x)

print("Похідна sin(x):", f1_der)
print("Похідна ln(x):", f2_der)

# Числові функції
f1_func = sp.lambdify(x, f1, 'numpy')
f1_der_func = sp.lambdify(x, f1_der, 'numpy')

f2_func = sp.lambdify(x, f2, 'numpy')
f2_der_func = sp.lambdify(x, f2_der, 'numpy')

# Діапазони значень
x_vals_1 = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_vals_2 = np.linspace(0.1, 10, 400)  # починаємо з 0.1, бо ln(0) — не існує

# Значення функцій
y1 = f1_func(x_vals_1)
y1_der = f1_der_func(x_vals_1)

y2 = f2_func(x_vals_2)
y2_der = f2_der_func(x_vals_2)

# Побудова графіків
plt.figure(figsize=(14, 6))

# Графік для sin(x)
plt.subplot(1, 2, 1)
plt.plot(x_vals_1, y1, label='sin(x)', linewidth=2)
plt.plot(x_vals_1, y1_der, label="cos(x)", linestyle='--')
plt.title('f(x) = sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.grid(True)
plt.legend()

# Графік для ln(x)
plt.subplot(1, 2, 2)
plt.plot(x_vals_2, y2, label='ln(x)', linewidth=2)
plt.plot(x_vals_2, y2_der, label="1/x", linestyle='--')
plt.title('f(x) = ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Оголошення символьної змінної
x = sp.Symbol('x')

# Задана функція витрат
f = x**2 - 5*x + 1

# Обчислення похідної
f_der = sp.diff(f, x)
print("Похідна функції:", f_der)

# Перетворення у числові функції
f_func = sp.lambdify(x, f, 'numpy')
f_der_func = sp.lambdify(x, f_der, 'numpy')

# Значення x на відрізку [0, 100]
x_vals = np.linspace(0, 100, 500)
y_vals = f_func(x_vals)
y_der_vals = f_der_func(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x² - 5x + 1', linewidth=2)
plt.plot(x_vals, y_der_vals, label="f'(x) = 2x - 5", linestyle='--')

plt.title('Функція витрат і її похідна')
plt.xlabel('Кількість одиниць (x)')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()
```
