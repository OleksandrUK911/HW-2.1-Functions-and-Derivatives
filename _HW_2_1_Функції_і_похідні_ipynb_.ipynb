### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# –°–∏–º–≤–æ–ª—å–Ω–∞ –∑–º—ñ–Ω–Ω–∞
x = sp.Symbol('x')

# –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è
f = 10 * x - 5

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –ø–æ—Ö—ñ–¥–Ω–æ—ó
f_derivative = sp.diff(f, x)
print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó:", f_derivative)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –Ω–∞ —á–∏—Å–ª–æ–≤—É —Ñ—É–Ω–∫—Ü—ñ—é
f_func = sp.lambdify(x, f, 'numpy')
f_der_func = sp.lambdify(x, f_derivative, 'numpy')

# –ó–Ω–∞—á–µ–Ω–Ω—è x –≤—ñ–¥ 0 –¥–æ 10
x_vals = np.linspace(0, 10, 200)
y_vals = f_func(x_vals)

# –ü–û–ü–†–ê–í–ö–ê: –ü–æ—Ö—ñ–¥–Ω–∞ ‚Äî —Å—Ç–∞–ª–∏–π –º–∞—Å–∏–≤
y_der_vals = np.full_like(x_vals, f_der_func(0))

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5', linewidth=2)
plt.plot(x_vals, y_der_vals, label="f'(x) = 10", linestyle='--')
plt.title('–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó')
plt.xlabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è (x)')
plt.ylabel('–û—Ü—ñ–Ω–∫–∞ / –®–≤–∏–¥–∫—ñ—Å—Ç—å –∑–º—ñ–Ω–∏ –æ—Ü—ñ–Ω–∫–∏')
plt.legend()
plt.grid(True)
plt.show()
```

–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó ‚Äî –ø—Ä—è–º–∞ –ª—ñ–Ω—ñ—è –≤–≥–æ—Ä—É, –∞ –≥—Ä–∞—Ñ—ñ–∫ –ø–æ—Ö—ñ–¥–Ω–æ—ó ‚Äî –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞ –ª—ñ–Ω—ñ—è.

–í–∏—Å–Ω–æ–≤–æ–∫: –ß–∏–º –±—ñ–ª—å—à–µ –Ω–∞–≤—á–∞—î—à—Å—è ‚Äî —Ç–∏–º –≤–∏—â–∞ –æ—Ü—ñ–Ω–∫–∞, —ñ –∫–æ–∂–Ω–∞ –≥–æ–¥–∏–Ω–∞ –¥–æ–¥–∞—î —Å—Ç—ñ–ª—å–∫–∏ –∂ –±–∞–ª—ñ–≤, —è–∫ —ñ –ø–æ–ø–µ—Ä–µ–¥–Ω—è.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# –°–∏–º–≤–æ–ª—å–Ω–∞ –∑–º—ñ–Ω–Ω–∞
x = sp.Symbol('x')

# –§—É–Ω–∫—Ü—ñ—ó
f1 = sp.sin(x)
f2 = sp.ln(x)

# –ü–æ—Ö—ñ–¥–Ω—ñ
f1_der = sp.diff(f1, x)
f2_der = sp.diff(f2, x)

print("–ü–æ—Ö—ñ–¥–Ω–∞ sin(x):", f1_der)
print("–ü–æ—Ö—ñ–¥–Ω–∞ ln(x):", f2_der)

# –ß–∏—Å–ª–æ–≤—ñ —Ñ—É–Ω–∫—Ü—ñ—ó
f1_func = sp.lambdify(x, f1, 'numpy')
f1_der_func = sp.lambdify(x, f1_der, 'numpy')

f2_func = sp.lambdify(x, f2, 'numpy')
f2_der_func = sp.lambdify(x, f2_der, 'numpy')

# –î—ñ–∞–ø–∞–∑–æ–Ω–∏ –∑–Ω–∞—á–µ–Ω—å
x_vals_1 = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_vals_2 = np.linspace(0.1, 10, 400)  # –ø–æ—á–∏–Ω–∞—î–º–æ –∑ 0.1, –±–æ ln(0) ‚Äî –Ω–µ —ñ—Å–Ω—É—î

# –ó–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ–π
y1 = f1_func(x_vals_1)
y1_der = f1_der_func(x_vals_1)

y2 = f2_func(x_vals_2)
y2_der = f2_der_func(x_vals_2)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.figure(figsize=(14, 6))

# –ì—Ä–∞—Ñ—ñ–∫ –¥–ª—è sin(x)
plt.subplot(1, 2, 1)
plt.plot(x_vals_1, y1, label='sin(x)', linewidth=2)
plt.plot(x_vals_1, y1_der, label="cos(x)", linestyle='--')
plt.title('f(x) = sin(x) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞')
plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.grid(True)
plt.legend()

# –ì—Ä–∞—Ñ—ñ–∫ –¥–ª—è ln(x)
plt.subplot(1, 2, 2)
plt.plot(x_vals_2, y2, label='ln(x)', linewidth=2)
plt.plot(x_vals_2, y2_der, label="1/x", linestyle='--')
plt.title('f(x) = ln(x) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞')
plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# –û–≥–æ–ª–æ—à–µ–Ω–Ω—è —Å–∏–º–≤–æ–ª—å–Ω–æ—ó –∑–º—ñ–Ω–Ω–æ—ó
x = sp.Symbol('x')

# –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏—Ç—Ä–∞—Ç
f = x**2 - 5*x + 1

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –ø–æ—Ö—ñ–¥–Ω–æ—ó
f_der = sp.diff(f, x)
print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó:", f_der)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —É —á–∏—Å–ª–æ–≤—ñ —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
f_der_func = sp.lambdify(x, f_der, 'numpy')

# –ó–Ω–∞—á–µ–Ω–Ω—è x –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É [0, 100]
x_vals = np.linspace(0, 100, 500)
y_vals = f_func(x_vals)
y_der_vals = f_der_func(x_vals)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x¬≤ - 5x + 1', linewidth=2)
plt.plot(x_vals, y_der_vals, label="f'(x) = 2x - 5", linestyle='--')

plt.title('–§—É–Ω–∫—Ü—ñ—è –≤–∏—Ç—Ä–∞—Ç —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞')
plt.xlabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –æ–¥–∏–Ω–∏—Ü—å (x)')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.legend()
plt.grid(True)
plt.show()
```
